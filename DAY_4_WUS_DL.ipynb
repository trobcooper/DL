```python
from sklearn.svm import LinearSVC
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
iris = datasets.load_iris()
features = iris.data[:100,:2]
target = iris.target[:100]
```

```python
target
```

```python
features
```

```python
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
```

```python
svc = LinearSVC(C=1.0)
svc.fit(features_scaled,target)
```

```python
modelm = svc.fit(features_scaled, target)
```

```python
modelm
```

```python
modelm.predict(features_scaled)
```

```python
modelm.predict(scaler.transform([[5.9, 3.0,]]))
```

```python
import matplotlib.pyplot as plt
color = ["black" if c == 0 else "blue" for c in target]
plt.scatter(features_scaled[:,0],  features_scaled[:,1],c=color)
w = svc.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(-2.5, 2.5)
yy = a * xx - (svc.intercept_[0]) / w[1]
plt.plot(xx, yy)
plt.axis("off"),plt.show();
```

```python
color = ["black" if c == 0 else "blue" for c in target]
plt.scatter(features_scaled[:,0],  features_scaled[:,1],c=color)
w = svc.coef_[0]
a = 1.1
xx = np.linspace(-2.5, 2.5)
yy = a * xx - (svc.intercept_[0]) / w[1]
plt.plot(xx, yy)
plt.axis("off"),plt.show();
```
